                        <h2 id="battleship">Battleship</h2>
<p>In this project you will build a simplified, one-player version of the classic board game Battleship! In this version of the game, there will be a single ship hidden in a random location on a 5x5 grid. The player will have 10 guesses to try to sink the ship.</p>
<p>To build this game we will use our knowledge of lists, conditionals and functions in Python.</p>
<p>We will go step by step, given solutions  at the end you will find the solution(complete code stored in <code>battleship.py</code>).</p>
<h4 id="ready-the-computing-starts-here">Ready? The computing starts here:</h4>
<ul>
<li><p>Create a variable board and set it equal to an empty list.</p>
</li>
<li><p>Create a 5 x 5 grid initialized to all &#39;O&#39;s and store it in board.
Tip:</p>
<pre><code class="lang-python"><span class="hljs-prompt">&gt;&gt;&gt; </span>board=[<span class="hljs-string">"O"</span>]*<span class="hljs-number">5</span>
<span class="hljs-prompt">&gt;&gt;&gt; </span><span class="hljs-keyword">print</span> board

<span class="hljs-prompt">&gt;&gt;&gt; </span><span class="hljs-keyword">print</span> len(board)
<span class="hljs-number">5</span>
`
</code></pre>
<ul>
<li>Use <code>range()</code> to loop 5 times.</li>
<li>Inside the loop, <code>.append()</code> a list containing 5 &quot;O&quot;s to board, just like in the example above.</li>
<li>Note that these are capital letter &quot;O&quot; and not zeros.</li>
<li>Use the <code>print</code> command to display the contents of the board list.</li>
</ul>
</li>
</ul>

<pre><code class="lang-python">board=[]

<ul>
<li>We can use the fact that our board is a list of lists to help us do this. Let&#39;s set up a for loop to go through each of the elements in the outer list (each of which is a row of our board) and print them.<ul>
<li>First, delete your existing print statement.
Then, define a function named <code>print_board</code> with a single argument, board.</li>
</ul>
</li>
<li>Inside the function, write a for loop to iterates through each row in board and print it to the screen.</li>
<li>Call your function with board to make sure it works.</li>
</ul>

<ul>
<li>Now we want to get rid of the commas, like this:<pre><code>&gt;&gt;&gt; letters = [&#39;a&#39;, &#39;b&#39;, &#39;c&#39;, &#39;d&#39;]
&gt;&gt;&gt; print &quot; &quot;.join(letters)
a b c d
&gt;&gt;&gt;
</code></pre><ul>
<li>Inside your function, inside your for loop, use &quot; &quot; as the separator to <code>.join</code> the elements of each row.</li>
</ul>
</li>
</ul>

print_board(board)

<ul>
<li><p>Now, let&#39;s hide our battleship in a random location on the board.Since we have a 2-dimensional list, we&#39;ll use two variables to store the ship&#39;s location, <code>ship_row</code> and <code>ship_col</code>.
Look at his example:</p>
<pre><code class="lang-python"><span class="hljs-prompt">&gt;&gt;&gt; </span><span class="hljs-keyword">from</span> random <span class="hljs-keyword">import</span> randint
<span class="hljs-prompt">&gt;&gt;&gt; </span>coin = randint(<span class="hljs-number">0</span>, <span class="hljs-number">1</span>)

&gt;&gt;&gt;
</code></pre>
<p>In the above example, we first import the <code>randint(low, high)</code> function from the random module.
Then, we generate either zero or one and store it in coin.
Finally, we generate a number from one to six inclusive.
Let&#39;s generate a <code>random_row</code> and <code>random_col</code> from zero to four.</p>
<ul>
<li>Define two new functions, <code>random_row</code> and <code>random_col</code>, that each take board as input.</li>
<li>These functions should return a random row index and a random column index from your board, respectively. Use randint(0, len(board) - 1).</li>
<li>Call each function on board.</li>
</ul>
</li>
</ul>
<p><strong>Solution 4</strong></p>
<pre><code class="lang-python"><span class="hljs-function"><span class="hljs-keyword">def</span> <span class="hljs-title">random_row</span><span class="hljs-params">(board)</span>:</span>
    <span class="hljs-keyword">from</span> random <span class="hljs-keyword">import</span> randint
    <span class="hljs-keyword">return</span> randint(<span class="hljs-number">0</span>, len(board) - <span class="hljs-number">1</span>)

<span class="hljs-function"><span class="hljs-keyword">def</span> <span class="hljs-title">random_col</span><span class="hljs-params">(board)</span>:</span>
    <span class="hljs-keyword">from</span> random <span class="hljs-keyword">import</span> randint
    <span class="hljs-keyword">return</span> randint(<span class="hljs-number">0</span>, len(board) - <span class="hljs-number">1</span>)
</code></pre>
<ul>
<li><p>We are going to ask the user to guess the column and the row where our ship is stored:</p>
<ul>
<li>Create a new variable called <code>guess_row and</code> set it to <code>int(raw_input(&quot;Guess Row: &quot;))</code>.</li>
<li>Create a new variable called <code>guess_col</code> and set it to <code>int(raw_input(&quot;Guess Col: &quot;))</code>.</li>
</ul>
</li>
<li><p>For now, while we&#39;re writing and debugging this part of the program, it will be helpful to know where that battleship is hidden. Let&#39;s add a print statement that displays the location of the hidden ship.We will delete it later on.</p>
</li>
<li>We have the actual location of the ship and the player&#39;s guess so we can check to see if the player guessed right.For a guess to be right, <code>guess_col</code> should be equal to <code>ship_col</code> and <code>guess_row</code> should be equal to <code>ship_row</code>.<ul>
<li>Add an else under the if &quot;correct condition&quot;.</li>
<li>Print out &quot;You missed my battleship!&quot;</li>
<li>Set the list element at <code>guess_row</code>, <code>guess_col</code> to &quot;X&quot;.</li>
<li>As the last line in your else statement, call <code>print_board(board)</code> again so you can see the &quot;X&quot;.</li>
</ul>
</li>
</ul>
<p><strong>Solution 5</strong></p>
<pre><code class="lang-python">ship_row = random_row(board)
ship_col = random_col(board)
guess_row = int(raw_input(<span class="hljs-string">"Guess Row:"</span>))
guess_col = int(raw_input(<span class="hljs-string">"Guess Col:"</span>))

<span class="hljs-keyword">print</span> ship_row
<span class="hljs-keyword">print</span> ship_col

<ul>
<li><p>Now let’s think a little bit more about the &quot;miss&quot; condition.</p>
<ul>
<li>The user can enter a guess that&#39;s off the board.</li>
<li>He can guess a spot they’ve already guessed.</li>
<li>He can just miss the ship.
We&#39;ll add the first case:</li>
<li>Add a new if: statement that is nested under the else.</li>
<li>It should check if <code>guess_row</code> is not in range(5) or <code>guess_col</code> is not in range(5).</li>
<li>If that is the case, print out &quot;Oops, that&#39;s not even in the ocean.&quot;</li>
<li>After your new if: statement, add an else: that contains your existing handler for an incorrect guess. Don&#39;t forget to indent the code.</li>
</ul>
</li>
<li><p>And now the second one:</p>
<ul>
<li>Add an elif to see if the guessed location already has an &#39;X&#39; in it.(board[col][row]==&quot;X&quot;)
If it has, print &quot;You guessed that one already.&quot;</li>
</ul>
</li>
</ul>
<p><strong>Solution 6</strong></p>
<pre><code class="lang-python"><span class="hljs-keyword">if</span> guess_row==ship_row <span class="hljs-keyword">and</span> guess_col==ship_col:
    <span class="hljs-keyword">print</span> <span class="hljs-string">"Congratulations!You sank my battleship!"</span>
<span class="hljs-keyword">else</span>:
         <span class="hljs-keyword">if</span> guess_col <span class="hljs-keyword">not</span> <span class="hljs-keyword">in</span> range(<span class="hljs-number">5</span>) <span class="hljs-keyword">or</span>  guess_row <span class="hljs-keyword">not</span> <span class="hljs-keyword">in</span>  range(<span class="hljs-number">5</span>):
           <span class="hljs-keyword">print</span> <span class="hljs-string">"Oops, that's not even in the ocean."</span>
         <span class="hljs-keyword">elif</span> board[guess_row][guess_col]==<span class="hljs-string">"X"</span>:
             <span class="hljs-keyword">print</span> <span class="hljs-string">"You guessed that one already."</span>

         <span class="hljs-keyword">else</span>:

            <span class="hljs-keyword">print</span> <span class="hljs-string">"You missed my battleship!"</span>
            board[guess_row][guess_col]=<span class="hljs-string">"X"</span>
            print_board(board)
</code></pre>
<ul>
<li><p>We’d like our game to allow the player to make up to 4 guesses before they lose.We can use a <code>for</code> loop to iterate through a range. Each iteration will be a turn.</p>
</li>
<li><p>Add a <code>for</code> loop that repeats the guessing and checking part of your game for 4 turns.</p>
</li>
<li>At the beginning of each iteration, print &quot;Turn&quot;, turn + 1 to let the player know what turn they are on.</li>
<li>Indent everything that should be repeated.</li>
</ul>
<p><strong>Solution 7</strong></p>
<pre><code class="lang-python"><span class="hljs-keyword">for</span> turn <span class="hljs-keyword">in</span> range(<span class="hljs-number">4</span>):


 guess_row = int(raw_input(<span class="hljs-string">"Guess Row:"</span>))

 <span class="hljs-keyword">if</span> guess_row == ship_row <span class="hljs-keyword">and</span> guess_col == ship_col:

 <span class="hljs-keyword">else</span>:
    <span class="hljs-keyword">if</span> (guess_row &lt; <span class="hljs-number">0</span> <span class="hljs-keyword">or</span> guess_row &gt; <span class="hljs-number">4</span>) <span class="hljs-keyword">or</span> (guess_col &lt; <span class="hljs-number">0</span> <span class="hljs-keyword">or</span>  guess_col &gt; <span class="hljs-number">4</span>):
        <span class="hljs-keyword">print</span> <span class="hljs-string">"Oops, that's not even in the ocean."</span>
    <span class="hljs-keyword">elif</span>(board[guess_row][guess_col] == <span class="hljs-string">"X"</span>):

    <span class="hljs-keyword">print</span> <span class="hljs-string">"Turn:"</span>,turn+<span class="hljs-number">1</span>
    print_board(board)

<ul>
<li>If someone runs out of guesses without winning right now, the game just exits. It would be nice to let them know why.Since we only want this message to display if the user guesses wrong on their last turn, we need to think carefully about where to put it.<ul>
<li>We’ll want to put it under the else that accounts for misses.</li>
</ul>
</li>
<li>We’ll want to print the message no matter what the cause of the miss.</li>
<li><p>Since our turn variable starts at 0 and goes to 3, we will want to end the game when turn equals 3.</p>
