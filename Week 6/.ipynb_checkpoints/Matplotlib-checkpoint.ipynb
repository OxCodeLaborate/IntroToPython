# What is matplotlib.pyplot (plt)?

````python
import matplotlib.pyplot as plt
````

````python
print(plt.__doc__)
````

What you *really* need to know is
```plt.plot(x, y)```


````python
import numpy as np
x = np.arange(10)
y = np.random.randint(0, 20, 10)
````

````python
plt.plot(x, y)
````

Of course we might want our plots to be labelled

````python
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Some plot')
plt.plot(x, y)
````

Or maybe we want dots on our plot instead of a line

````python
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Some plot')
plt.plot(x, y, '.')
````

Want some color, anyone?

You can also plot multiple things on the same plot

````python
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Some plot')
plt.plot(x, y, 'r.',
         x, x + y, 'g.'
        )
````
